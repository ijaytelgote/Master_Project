```python
import pandas as pd
df=pd.read_csv('/content/winequalityN.csv')
df.sample(5)
```

```python
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import numpy as np
```

```python

# Create a scatter plot of the original data
plt.figure(figsize=(10, 5))
plt.scatter(df['sulphates'], df['alcohol'], c='yello', label='Normal')
plt.title('Original Data')
plt.legend()
plt.show()
```

```python
data=df[['alcohol','sulphates']]
```

```python
data.isnull().sum()
```

```python
import pandas as pd

# Assuming df is your DataFrame
data.dropna(inplace=True)
data.head()
```

```python
d=data.values.tolist()
```

```python
# Assuming you have a DataFrame named 'wine_data' with columns representing different features including 'quality'

# Import necessary libraries

# Create and fit the Isolation Forest model
clf = IsolationForest(contamination=0.05)  # Adjust the contamination parameter based on your dataset
clf.fit(d)

# Predict outliers (1 for inliers, -1 for outliers)
predictions = clf.predict(d)

# Visualize the data and outliers
plt.scatter(data['alcohol'], data['sulphates'], c=predictions, cmap='viridis')
plt.title('Isolation Forest Anomaly Detection in Wine Qualities')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Outlier Prediction')
plt.show()
```
