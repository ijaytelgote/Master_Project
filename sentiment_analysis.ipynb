```python
df=pd.read_csv('/content/Real_CRM_Data.csv')
df.head(2)
```

```python
df=df.rename(columns={'Unnamed: 0':'Customer ID'})
df.head()
```

```python
#to encounter negative words
import pandas as pd
import plotly.express as px
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import download

# Download the VADER lexicon for sentiment analysis
download('vader_lexicon')

# Placeholder DataFrame with a 'Communication History' column

# Sentiment analysis
sid = SentimentIntensityAnalyzer()
df['compound'] = df['Communication History'].apply(lambda x: sid.polarity_scores(x)['compound'])

# Classify sentiments
df['sentiment'] = df['compound'].apply(lambda x: 'positive' if x >= 0.05 else 'negative' if x <= -0.05 else 'neutral')

# Visualize sentiments using Plotly
fig = px.bar(df, x='sentiment', title='Sentiment Distribution',
             labels={'sentiment': 'Sentiment', 'index': 'Count'},
             color='sentiment',
             color_discrete_map={'positive': 'green', 'negative': 'red', 'neutral': 'gray'},
             hover_data=['Customer ID'])

fig.show()
```
